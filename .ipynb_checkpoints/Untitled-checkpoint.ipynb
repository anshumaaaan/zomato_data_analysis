```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato-data-.csv")
print(dataframe.head())
```

```python
def handleRate(value):
    value = str(value).split('/')
    value=value[0];
    return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])  # or try 'pastel', 'husl', etc.
plt.xlabel("Type of restaurant")
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes' : grouped_data})
plt.plot(result, c='green', marker = 'o')
plt.xlabel("Type of restaurant", c='red', size = 10)
plt.ylabel('Votes', c = "red", size = 10)
```

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']

print("Restaurant(s) with the maximum votes: ")
print(restaurant_with_max_votes)
```

```python
sns.countplot(x=dataframe['online_order'])
```

```python
plt.hist(dataframe['rate'], bins = 5)
plt.title('Ratings Distribution')
plt.show()
```

```python
# approximate cost for two people
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
plt.xticks(rotation = 45)
most_common_cost = dataframe['approx_cost(for two people)'].mode()[0]
print("Most preferred cost for two people:", most_common_cost)
```

```python
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```

```python
# Dining restaurants primarily accept offline orders whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurants but prefer online ordering at cafes.
```
